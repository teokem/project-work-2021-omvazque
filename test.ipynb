```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
from glob import glob
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
year = ['2020', '2021']
month = ['01','02','03','04','05','06','07','08','09','10','11','12']
days_2020 = [31,29,31,30,31,30,31,31,30,31,30,31]
days_2021 = [31,28,31,30,31,30,31,31,30,31,30,31]

start_day = 1
start_month = '04'

stop_day = 31
stop_month = '03'

# Data will be used from the 1st of April 2020 until 31st of March 2021

#dates = []
apr_2020 = []
may_2020 = []
jun_2020 = []
jul_2020 = []
aug_2020 = []
sep_2020 = []
oct_2020 = []
nov_2020 = []
dec_2020 = []
jan_2021 = []
feb_2021 = []
mar_2021 = []

# arrays per year/month
                            
for d in range(days_2020[month.index('04')]):
    if(d > 10):
        apr_2020.append("{}-{}-{}".format('04',d+1,'2020'))

for d in range(days_2020[month.index('05')]):
    if(d < 9):
        may_2020.append("{}-0{}-{}".format('05',d+1,'2020'))
    else:
        may_2020.append("{}-{}-{}".format('05',d+1,'2020'))
        
for d in range(days_2020[month.index('06')]):
    if(d < 9):
        jun_2020.append("{}-0{}-{}".format('06',d+1,'2020'))
    else:
        jun_2020.append("{}-{}-{}".format('06',d+1,'2020'))
        
for d in range(days_2020[month.index('07')]):
    if(d < 9):
        jul_2020.append("{}-0{}-{}".format('07',d+1,'2020'))
    else:
        jul_2020.append("{}-{}-{}".format('07',d+1,'2020'))
        
for d in range(days_2020[month.index('08')]):
    if(d < 9):
        aug_2020.append("{}-0{}-{}".format('08',d+1,'2020'))
    else:
        aug_2020.append("{}-{}-{}".format('08',d+1,'2020'))
    
for d in range(days_2020[month.index('09')]):
    if(d < 9):
        sep_2020.append("{}-0{}-{}".format('09',d+1,'2020'))
    else:
        sep_2020.append("{}-{}-{}".format('09',d+1,'2020'))
        
for d in range(days_2020[month.index('10')]):
    if(d < 9):
        oct_2020.append("{}-0{}-{}".format('10',d+1,'2020'))
    else:
        oct_2020.append("{}-{}-{}".format('10',d+1,'2020'))
        
for d in range(days_2020[month.index('11')]):
    if(d < 9):
        nov_2020.append("{}-0{}-{}".format('11',d+1,'2020'))
    else:
        nov_2020.append("{}-{}-{}".format('11',d+1,'2020'))
        
for d in range(days_2020[month.index('12')]):
    if(d < 9):
        dec_2020.append("{}-0{}-{}".format('12',d+1,'2020'))
    else:
        dec_2020.append("{}-{}-{}".format('12',d+1,'2020'))

for d in range(days_2021[month.index('01')]):
    if(d < 9):
        jan_2021.append("{}-0{}-{}".format('01',d+1,'2021'))
    else:
        jan_2021.append("{}-{}-{}".format('01',d+1,'2021'))

for d in range(days_2021[month.index('02')]):
    if(d < 9):
        feb_2021.append("{}-0{}-{}".format('02',d+1,'2021'))
    else:
        feb_2021.append("{}-{}-{}".format('02',d+1,'2021'))
        
for d in range(days_2021[month.index('03')]):
    if(d>stop_day):
        break
    if(d < 9):
        mar_2021.append("{}-0{}-{}".format('03',d+1,'2021'))
    else:
        mar_2021.append("{}-{}-{}".format('03',d+1,'2021'))
    
print(mar_2021)
print(len(mar_2021))
```

```python
frame_apr_2020 = []
frame_may_2020 = []
frame_jun_2020 = []
frame_jul_2020 = []
frame_aug_2020 = []
frame_sep_2020 = []
frame_oct_2020 = []
frame_nov_2020 = []
frame_dec_2020 = []
frame_jan_2021 = []
frame_feb_2021 = []
frame_mar_2021 = []

path = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports_us'
Columns = ['Province_State','Country_Region','Last_Update','Lat','Long_','Confirmed','Deaths','Recovered','Active','FIPS','Incident_Rate','People_Tested','People_Hospitalized','Mortality_Rate']
Names = ['Province_State','Country_Region','Last_Update','Lat','Long_','Confirmed','Deaths','Recovered','Active','FIPS','Incident_Rate','People_Tested','People_Hospitalized','Mortality_Rate']

for f in apr_2020:
    frame_apr_2020.append(pd.read_csv('{}/{}.csv'.format(path,f),header=0,usecols=Columns,names=Names))

for f in may_2020:
    frame_may_2020.append(pd.read_csv('{}/{}.csv'.format(path,f),header=0,usecols=Columns,names=Names))

for f in jun_2020:
    frame_jun_2020.append(pd.read_csv('{}/{}.csv'.format(path,f),header=0,usecols=Columns,names=Names))

for f in jul_2020:
    frame_jul_2020.append(pd.read_csv('{}/{}.csv'.format(path,f),header=0,usecols=Columns,names=Names))

for f in aug_2020:
    frame_aug_2020.append(pd.read_csv('{}/{}.csv'.format(path,f),header=0,usecols=Columns,names=Names))
    
for f in sep_2020:
    frame_sep_2020.append(pd.read_csv('{}/{}.csv'.format(path,f),header=0,usecols=Columns,names=Names))
    
for f in oct_2020:
    frame_oct_2020.append(pd.read_csv('{}/{}.csv'.format(path,f),header=0,usecols=Columns,names=Names))
    
for f in nov_2020:
    frame_nov_2020.append(pd.read_csv('{}/{}.csv'.format(path,f),header=0,usecols=Columns,names=Names))
    
for f in dec_2020:
    frame_dec_2020.append(pd.read_csv('{}/{}.csv'.format(path,f),header=0,usecols=Columns,names=Names))
    
for f in jan_2021:
    frame_jan_2021.append(pd.read_csv('{}/{}.csv'.format(path,f),header=0,usecols=Columns,names=Names))
    
for f in feb_2021:
    frame_feb_2021.append(pd.read_csv('{}/{}.csv'.format(path,f),header=0,usecols=Columns,names=Names))
    
for f in mar_2021:
    frame_mar_2021.append(pd.read_csv('{}/{}.csv'.format(path,f),header=0,usecols=Columns,names=Names))
```

```python
# The 'pd.crosstab' funtion is used to normalize the data. 
# The bars represent the percetage of Confirmed cases per State 
# until the last day of available data (31 March 2021) 

variable_to_plot = 'Confirmed'
co_total = pd.crosstab(index=frame_mar_2021[-1].Province_State,
                       columns=frame_mar_2021[-1]['Country_Region'],values=frame_mar_2021[-1][variable_to_plot],
                       aggfunc='sum',normalize='columns',margins=True,margins_name='Confirmed')

fig = plt.figure(figsize=(30,10))
ax = fig.add_axes([0,0,1,1])

plt.title("Percentage of {} cases of COVID-19 per State until March 31, 2021".format(variable_to_plot),fontdict={'fontsize': 30})
plt.xlabel("Sates",fontdict={'fontsize': 30})
plt.ylabel("Percentage",fontdict={'fontsize': 30})
plt.xticks(rotation=90)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

ax.bar(co_total.Confirmed.index.values,co_total.Confirmed.values)
plt.show()
```

```python
# The 'pd.crosstab' funtion is used to normalize the data. 
# Here I represent in a pie chart the top 5 States with Confirmed cases
# until the last day of available data (31 March 2021) 

plt.style.use('fivethirtyeight')
variable_to_plot = 'Confirmed'
co_total = pd.crosstab(index=frame_mar_2021[-1].Province_State,
                       columns=frame_mar_2021[-1]['Country_Region'],values=frame_mar_2021[-1][variable_to_plot],
                       aggfunc='sum',normalize='columns',margins=True,margins_name='Confirmed')


x = co_total.Confirmed.index.values
y = co_total.Confirmed.values

n = 5
top_values = sorted(y)[-n:]
top_indexes = []
indexes = []
    
for i in top_values:
    top_indexes.append(co_total.index[co_total['Confirmed'] == i].tolist())
    
for i in range(n):
    indexes.append(top_indexes[i][0])

top_values.append(1.0-(np.array(top_values)).sum())
indexes.append('Rest')    

fig1, ax1 = plt.subplots()
ax1.pie(x=top_values, labels=indexes,
        autopct='%1.1f%%',shadow=False, startangle=90,normalize=False)
ax1.axis('equal')
plt.title('Top {} states with {} cases of COVID-19'.format(n,variable_to_plot))
plt.figure(figsize=(50,50))
plt.rcParams["figure.figsize"]=(6, 5)
plt.show()
```
