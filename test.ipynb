## Abstract of project

In the [JHUCSSE](https://github.com/CSSEGISandData) repository one can find daily reports of the number of Confirmed, Deaths, Recovered, etc. cases. For example, one can find the file `04-12-2020.csv` associated to report from the 12th of April of 2020. In this notebook, I am interested in the Confirmed and Deaths cases as a function of the state. In addition only data dating from the 12th of April 2020 to the 31st of March 2021 will be taken into account.

## Import required packages 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
from glob import glob
import matplotlib.pyplot as plt
%matplotlib inline
```

One efficient tool to mage large amount of data is the **pandas** library. The code below declares lists associated to different months (`frame_apr_2020, frame_may_2020, etc.`). The `pd.read_csv()` function can be used to read and store data frames into the previously defined lists.

```python
path = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports_us'
Columns = ['Province_State','Country_Region','Last_Update','Lat','Long_','Confirmed','Deaths','Recovered','Active','FIPS','Incident_Rate','People_Tested','People_Hospitalized','Mortality_Rate']
Names = ['Province_State','Country_Region','Last_Update','Lat','Long_','Confirmed','Deaths','Recovered','Active','FIPS','Incident_Rate','People_Tested','People_Hospitalized','Mortality_Rate']

year = ['2020', '2021']
month = ['01','02','03','04','05','06','07','08','09','10','11','12']
days_2020 = [31,29,31,30,31,30,31,31,30,31,30,31]
days_2021 = [31,28,31,30,31,30,31,31,30,31,30,31]

start_day = 10

frame_apr_2020 = []
frame_may_2020 = []
frame_jun_2020 = []
frame_jul_2020 = []
frame_aug_2020 = []
frame_sep_2020 = []
frame_oct_2020 = []
frame_nov_2020 = []
frame_dec_2020 = []
frame_jan_2021 = []
frame_feb_2021 = []
frame_mar_2021 = []

for d in range(days_2020[month.index('04')]):
    if(d > start_day):
        frame_apr_2020.append(pd.read_csv('{}/{}-{}-{}.csv'.format(path,'04',d+1,'2020'),header=0,usecols=Columns,names=Names))

for d in range(days_2020[month.index('05')]):
    if(d < 9):
        frame_may_2020.append(pd.read_csv('{}/{}-0{}-{}.csv'.format(path,'05',d+1,'2020'),header=0,usecols=Columns,names=Names))
    else:
        frame_may_2020.append(pd.read_csv('{}/{}-{}-{}.csv'.format(path,'05',d+1,'2020'),header=0,usecols=Columns,names=Names))
        
for d in range(days_2020[month.index('06')]):
    if(d < 9):
        frame_jun_2020.append(pd.read_csv('{}/{}-0{}-{}.csv'.format(path,'06',d+1,'2020'),header=0,usecols=Columns,names=Names))
    else:
        frame_jun_2020.append(pd.read_csv('{}/{}-{}-{}.csv'.format(path,'06',d+1,'2020'),header=0,usecols=Columns,names=Names))

for d in range(days_2020[month.index('07')]):
    if(d < 9):
        frame_jul_2020.append(pd.read_csv('{}/{}-0{}-{}.csv'.format(path,'07',d+1,'2020'),header=0,usecols=Columns,names=Names))
    else:
        frame_jul_2020.append(pd.read_csv('{}/{}-{}-{}.csv'.format(path,'07',d+1,'2020'),header=0,usecols=Columns,names=Names))

for d in range(days_2020[month.index('08')]):
    if(d < 9):
        frame_aug_2020.append(pd.read_csv('{}/{}-0{}-{}.csv'.format(path,'08',d+1,'2020'),header=0,usecols=Columns,names=Names))
    else:
        frame_aug_2020.append(pd.read_csv('{}/{}-{}-{}.csv'.format(path,'08',d+1,'2020'),header=0,usecols=Columns,names=Names))

for d in range(days_2020[month.index('09')]):
    if(d < 9):
        frame_sep_2020.append(pd.read_csv('{}/{}-0{}-{}.csv'.format(path,'09',d+1,'2020'),header=0,usecols=Columns,names=Names))
    else:
        frame_sep_2020.append(pd.read_csv('{}/{}-{}-{}.csv'.format(path,'09',d+1,'2020'),header=0,usecols=Columns,names=Names))

for d in range(days_2020[month.index('10')]):
    if(d < 9):
        frame_oct_2020.append(pd.read_csv('{}/{}-0{}-{}.csv'.format(path,'10',d+1,'2020'),header=0,usecols=Columns,names=Names))
    else:
        frame_oct_2020.append(pd.read_csv('{}/{}-{}-{}.csv'.format(path,'10',d+1,'2020'),header=0,usecols=Columns,names=Names))

for d in range(days_2020[month.index('11')]):
    if(d < 9):
        frame_nov_2020.append(pd.read_csv('{}/{}-0{}-{}.csv'.format(path,'11',d+1,'2020'),header=0,usecols=Columns,names=Names))
    else:
        frame_nov_2020.append(pd.read_csv('{}/{}-{}-{}.csv'.format(path,'11',d+1,'2020'),header=0,usecols=Columns,names=Names))

for d in range(days_2020[month.index('12')]):
    if(d < 9):
        frame_dec_2020.append(pd.read_csv('{}/{}-0{}-{}.csv'.format(path,'12',d+1,'2020'),header=0,usecols=Columns,names=Names))
    else:
        frame_dec_2020.append(pd.read_csv('{}/{}-{}-{}.csv'.format(path,'12',d+1,'2020'),header=0,usecols=Columns,names=Names))

for d in range(days_2021[month.index('01')]):
    if(d < 9):
        frame_jan_2021.append(pd.read_csv('{}/{}-0{}-{}.csv'.format(path,'01',d+1,'2021'),header=0,usecols=Columns,names=Names))
    else:
        frame_jan_2021.append(pd.read_csv('{}/{}-{}-{}.csv'.format(path,'01',d+1,'2021'),header=0,usecols=Columns,names=Names))
        
for d in range(days_2021[month.index('02')]):
    if(d < 9):
        frame_feb_2021.append(pd.read_csv('{}/{}-0{}-{}.csv'.format(path,'02',d+1,'2021'),header=0,usecols=Columns,names=Names))
    else:
        frame_feb_2021.append(pd.read_csv('{}/{}-{}-{}.csv'.format(path,'02',d+1,'2021'),header=0,usecols=Columns,names=Names))

for d in range(days_2021[month.index('03')]):
    if(d < 9):
        frame_mar_2021.append(pd.read_csv('{}/{}-0{}-{}.csv'.format(path,'03',d+1,'2021'),header=0,usecols=Columns,names=Names))
    else:
        frame_mar_2021.append(pd.read_csv('{}/{}-{}-{}.csv'.format(path,'03',d+1,'2021'),header=0,usecols=Columns,names=Names))
```

By running `frame_apr_2021[11].head()` in a code cell one can see the different type of information in the data frame.

I am interested in the number of **Confirmed** and **Deaths** cases as a function of the different states (Texas, California, Florida, etc.).
In the code below one can use the `pd.pivot_table` function to select only data of our interest. For instance the `Province_State` can be used as the row index and the values in the columns will correspond to the **Confirmed** and **Deaths** cases.
The number of cases up to a particular month is taken from its last day. For example `frame_jul_2020[-1]` will return the number of cases until the 30th of July of 2020.

```python
columns_names = ['apr_2020', 'may_2020','jun_2020', 'jul_2020','aug_2020', 'sep_2020','oct_2020', 'nov_2020','dec_2020', 'jan_2021', 'feb_2021','mar_2021']
Values = ['Confirmed','Deaths','Recovered','Active']

pivot_apr_2020 = pd.pivot_table(frame_apr_2020[-1],values=Values,index=['Province_State'],aggfunc='sum')
pivot_may_2020 = pd.pivot_table(frame_may_2020[-1],values=Values,index=['Province_State'],aggfunc='sum')
pivot_jun_2020 = pd.pivot_table(frame_jun_2020[-1],values=Values,index=['Province_State'],aggfunc='sum')
pivot_jul_2020 = pd.pivot_table(frame_jul_2020[-1],values=Values,index=['Province_State'],aggfunc='sum')
pivot_aug_2020 = pd.pivot_table(frame_aug_2020[-1],values=Values,index=['Province_State'],aggfunc='sum')
pivot_sep_2020 = pd.pivot_table(frame_sep_2020[-1],values=Values,index=['Province_State'],aggfunc='sum')
pivot_oct_2020 = pd.pivot_table(frame_oct_2020[-1],values=Values,index=['Province_State'],aggfunc='sum')
pivot_nov_2020 = pd.pivot_table(frame_nov_2020[-1],values=Values,index=['Province_State'],aggfunc='sum')
pivot_dec_2020 = pd.pivot_table(frame_dec_2020[-1],values=Values,index=['Province_State'],aggfunc='sum')
pivot_jan_2021 = pd.pivot_table(frame_jan_2021[-1],values=Values,index=['Province_State'],aggfunc='sum')
pivot_feb_2021 = pd.pivot_table(frame_feb_2021[-1],values=Values,index=['Province_State'],aggfunc='sum')
pivot_mar_2021 = pd.pivot_table(frame_mar_2021[-1],values=Values,index=['Province_State'],aggfunc='sum')
```

In the code below the aggregated number of Confirmed cases is plotted as a function of the state. One can see that until the 31st of March of 2021 California is the state with most of the Confirmed cases. In addition, in the period between November 2020 to January 2021 the majority of the new cases were reported.

```python
variable_to_plot = 'Confirmed'
fig = plt.figure(figsize=(30,10))
ax = plt.axes()
plt.title("{} cases of COVID-19 in different states".format(variable_to_plot),fontdict={'fontsize': 30})
plt.xlabel("Sates",fontdict={'fontsize': 30})
plt.ylabel("{} cases".format(variable_to_plot),fontdict={'fontsize': 30})
plt.xticks(rotation=90)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

ax.plot(pivot_apr_2020.index.values,pivot_apr_2020[variable_to_plot].values,marker='o',linestyle="None",label='Apr 2020')
ax.plot(pivot_may_2020.index.values,pivot_may_2020[variable_to_plot].values,marker='v',linestyle="None",label='May 2020')
ax.plot(pivot_jun_2020.index.values,pivot_jun_2020[variable_to_plot].values,marker='<',linestyle="None",label='June 2020')
ax.plot(pivot_jul_2020.index.values,pivot_jul_2020[variable_to_plot].values,marker='>',linestyle="None",label='July 2020')
ax.plot(pivot_aug_2020.index.values,pivot_aug_2020[variable_to_plot].values,marker='^',linestyle="None",label='August 2020')
ax.plot(pivot_sep_2020.index.values,pivot_sep_2020[variable_to_plot].values,marker='8',linestyle="None",label='September 2020')
ax.plot(pivot_oct_2020.index.values,pivot_oct_2020[variable_to_plot].values,marker='s',linestyle="None",label='October 2020')
ax.plot(pivot_nov_2020.index.values,pivot_nov_2020[variable_to_plot].values,marker='p',linestyle="None",label='November 2020')
ax.plot(pivot_dec_2020.index.values,pivot_dec_2020[variable_to_plot].values,marker='P',linestyle="None",label='December 2020')
ax.plot(pivot_jan_2021.index.values,pivot_jan_2021[variable_to_plot].values,marker='*',linestyle="None",label='January 2021')
ax.plot(pivot_feb_2021.index.values,pivot_feb_2021[variable_to_plot].values,marker='D',linestyle="None",label='February 2021')
ax.plot(pivot_mar_2021.index.values,pivot_mar_2021[variable_to_plot].values,marker='h',linestyle="None",label='March 2021')
ax.legend(prop={"size":20},ncol=3,loc='upper center')
```

In the code below the `pd.crosstab` function is used to normalize the data and to plot the percetage of Confirmed cases as a function of the state until the 31st of March of 2021. It is observed that California leads the number of Confirmed cases with more that 10%.

```python
variable_to_plot = 'Confirmed'
co_total = pd.crosstab(index=frame_mar_2021[-1].Province_State,
                       columns=frame_mar_2021[-1]['Country_Region'],values=frame_mar_2021[-1][variable_to_plot],
                       aggfunc='sum',normalize='columns',margins=True,margins_name='Confirmed')

fig = plt.figure(figsize=(30,10))
ax = fig.add_axes([0,0,1,1])

plt.title("Percentage of {} cases of COVID-19 per State until March 31, 2021".format(variable_to_plot),fontdict={'fontsize': 30})
plt.xlabel("Sates",fontdict={'fontsize': 30})
plt.ylabel("Percentage",fontdict={'fontsize': 30})
plt.xticks(rotation=90)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

ax.bar(co_total.Confirmed.index.values,co_total.Confirmed.values)
plt.show()
```

In the code below a pie chart is used to represent the top 5 states with Confirmed cases. California, Texas, Florida, New York and Illinois lead this category. The total remaining is represented as the Rest. 

```python
plt.style.use('fivethirtyeight')
variable_to_plot = 'Confirmed'
co_total = pd.crosstab(index=frame_mar_2021[-1].Province_State,
                       columns=frame_mar_2021[-1]['Country_Region'],values=frame_mar_2021[-1][variable_to_plot],
                       aggfunc='sum',normalize='columns',margins=True,margins_name='Confirmed')


x = co_total.Confirmed.index.values
y = co_total.Confirmed.values

n = 5
top_values = sorted(y)[-n:]
top_indexes = []
indexes = []
    
for i in top_values:
    top_indexes.append(co_total.index[co_total['Confirmed'] == i].tolist())
    
for i in range(n):
    indexes.append(top_indexes[i][0])

top_values.append(1.0-(np.array(top_values)).sum())
indexes.append('Rest')    

fig1, ax1 = plt.subplots()
ax1.pie(x=top_values, labels=indexes,
        autopct='%1.1f%%',shadow=False, startangle=90,normalize=False)
ax1.axis('equal')
plt.title('Top {} states with {} cases of COVID-19 until the 31st of March, 2021'.format(n,variable_to_plot))
plt.figure(figsize=(50,50))
plt.rcParams["figure.figsize"]=(6, 5)
plt.show()
```
